```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Create the dataset
data = {
    "Index": list(range(1, 51)),  # Add numbering from 1 to 50
    "pIC50": [
        5.299, 5.008, 5.267, 4.678, 5.005, 5.254, 5.224, 5.749, 5.463, 5.138,
        5.265, 5.452, 4.421, 5.709, 5.721, 5.488, 5.607, 5.82, 5.673, 5.869,
        5.707, 5.694, 5.603, 5.551, 5.982, 5.563, 5.542, 5.32, 5.57, 5.634,
        5.896, 5.744, 4.946, 5.329, 5.705, 5.559, 5.525, 4.429, 4.617, 5.496,
        4.814, 4.584, 4.996, 4.903, 4.966, 5.039, 5.531, 5.277, 5.787, 5.826
    ]
}

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Define the bins for stratification
# Group pIC50 values into bins to approximate equal representation in train/test split
bins = pd.cut(df["pIC50"], bins=5, labels=False)

# Split the data into training and testing sets (80:20) with stratification
train_data, test_data = train_test_split(df, test_size=0.2, stratify=bins, random_state=42)

# Output the results
print("Training Data:")
print(train_data.sort_values(by="Index"))  # Sort by Index for easier tracking
print("\nTesting Data:")
print(test_data.sort_values(by="Index"))  # Sort by Index for easier tracking
```
